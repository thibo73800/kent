# Linear regression


```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Number of data points
X = 200

# You can test your own formula by adding one here
formulas = [
    lambda x: 2*x**2 + 10 + (random.uniform(-1, 1)*x**2),
    lambda x: -2*x**2 + 10 + (random.uniform(-1, 1)*x**2),
    lambda x: x + (random.uniform(-1, 1)*x**2),
    lambda x: (x-100) ** 2 + (random.uniform(-50, 50)*abs(x - 100)),
]

def get_data(formula, N):
    """
        **input: **
            *formula: lambda formula to used
            *N (integer) number of points
    """
    X = np.array([x for x in range(0, N)]) # Generate X points
    Y = np.array([formula(x)  for x in range(0, N)]) # Generate Y points 
   
    return X, Y

# By default used the first formula
X, Y = get_data(formulas[0], 200)
```

# Plot data - [First formula ]

```python
plt.plot(X, Y, '.')
plt.title("Y from X ")
plt.show() 
```

## Explore the relationship: Covariance
<img src="images/covariance.png"/>

### Mean X

```python
x_mean_line = [np.mean(X)] * N
y_mean_line = [np.mean(Y)] * N

plt.plot(X, Y, '.')
plt.plot(x_mean_line, Y, '-')
plt.title("X mean")
plt.show()
```

### Mean Y

```python
plt.plot(X, Y, '.')
plt.plot(X, y_mean_line, '-')
plt.title("Y mean")
plt.show()
```

### Cov(x, y)

```python
covariance = np.sum((X - np.mean(X))*(Y - np.mean(Y))) / (N - 1)
print("covariance=", covariance)
```

# Compute r by hand

<img src="images/compute_r.png" />

Measures the relative strength of the linear
relationship between two variables
*Unit-less
*Ranges between –1 and 1

*The closer to –1, the stronger the negative linear relationship

*The closer to 1, the stronger the positive linear relationship

*The closer to 0, the weaker any positive linear relationship


```python
# Variance of X
var_x = np.sum(((X - np.mean(X)) ** 2) / (N - 1))
# Variance of Y
var_y = np.sum(((Y - np.mean(Y)) ** 2) / (N - 1))

print("Variance X", var_x)
print("Variance Y", var_y)

# Compute the Pearson-s correlation coefficient
r = covariance / (np.sqrt(var_x) * np.sqrt(var_y))

print("Pearsons correlation coefficient: r =", r)
```

## Compute the slope β and the intercept α

<img src="images/compute.png" />

```python
# Take the first formula
X, Y = get_data(formulas[0], 200)

# Comptue covariance
covariance = np.sum((X - np.mean(X))*(Y - np.mean(Y)))
# Variance of X
var_x = np.sum((X - np.mean(X)) ** 2)


# Comput the slope using the formula above
β = covariance / var_x
# Compute the intercept using the formula above
α = np.mean(Y) - (β*np.mean(X))

# Set the predicted formula and predicted points
predicted_equation = lambda x: x*β + α
predictions = [predicted_equation(x) for x in range(N)]


# Plot function
plt.plot(X, Y, '.')
plt.plot(X, predictions, '-')
plt.title("Y from X and the predicted line")
plt.show() 
```

### Compute the errors of the regression line

<img src="images/errors.png" />


```python
SSreg = np.sum((predictions - np.mean(Y))**2)
SSridisual = np.sum((predictions - Y)**2)

SStotal = SSreg + SSridisual
R2 = SSreg / SStotal

print("SSridisual (Sum of the Square Error - SSE) =", SSridisual)
print("SStotal = ", SStotal)

# R2 Sould be the closest to 1 as possible
print("R2", R2)
```

## Compute the slope β and the intercept α for each formula

```python
for formula in formulas:
    ### # Take the first formula
    X, Y = get_data(formula, 200)

    # Comptue covariance
    covariance = np.sum((X - np.mean(X))*(Y - np.mean(Y)))
    # Variance of X
    var_x = np.sum((X - np.mean(X)) ** 2)


    # Comput the slope using the formula above
    β = covariance / var_x
    # Compute the intercept using the formula above
    α = np.mean(Y) - (β*np.mean(X))


    # Set the predicted formula and predicted points
    predicted_equation = lambda x: x*β + α
    predictions = [predicted_equation(x) for x in range(N)]

    # Compute errors
    SSreg = np.sum((predictions - np.mean(Y))**2)
    SSridisual = np.sum((predictions - Y)**2)

    SStotal = SSreg + SSridisual
    R2 = SSreg / SStotal

    # Plot function
    plt.plot(X, Y, '.')
    plt.plot(X, predictions, '-')
    plt.title("Y from X and the predicted line")
    plt.show() 
    

    print("SSridisual (Sum of the Square Error - SSE) =", SSridisual)
    print("SStotal = ", SStotal)
    # R2 Sould be the closest to 1 as possible
    print("R2", R2)
```
